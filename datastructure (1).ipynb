Q1. What are data structures, and why are they important?
   Data structures are ways of organizing and storing data in a computer so that it can be accessed and modified efficiently. Examples include lists, tuple, and dictionaries. They are important because they allow efficient use of memory, faster data access, and better problem solving techniques.

Q2. Explain the difference between mutable and immutable data types with examples.
   - Mutable: Can be changed after creating. Example: list = [1,2,3], we can append or remove elements.
   - Immutable: Cannot be changed after creating. Example: tuple = (1,2,3) or string = 'hello'.

Q3. What are the main differences between lists and tuples in Python?
   Lists are mutable; tuples are immutable. Lists use []; tuples use (). Tuples are often used for fixed data.

Q4. Describe how dictionaries store data.
   Dictionaries store data in key and value pairs. Each key must be unique or the last same key will be taken for the value, while values can be of any data type. Internally, Python dictionaries use a hash table to provide fast access.

Q5. Why might you use a set instead of a list in Python?
   Sets automatically remove duplicates. Useful when order doesn’t matter and uniqueness is required.

Q6. What is a string in Python, and how is it different from a list?
   A string is a sequence of characters enclosed in quotes. It is immutable. A list is a collection of elements (numbers, strings, etc.) enclosed in [ ] and is mutable.

Q7. How do tuples ensure data integrity in Python?
   Tuples are immutable, meaning once created, their contents cannot be changed. This prevents accidental modification, ensuring data integrity.

Q8. What is a hash table, and how does it relate to dictionaries in Python?
  A hash table is a data structure that stores data in an array like format using hash functions to map keys to indices. Python dictionaries use key and values and use unhashable content like tuples.

Q9. Can lists contain different data types in Python?
  Yes, lists can hold elements of different types, e.g., [1, 'hello', 3.14, True].

Q10. Explain why strings are immutable in Python.
    Strings are immutable because each modification creates a new string object.

Q11. What advantages do dictionaries offer over lists for certain tasks?
  - Faster for finding values assigned to specific  keys.
  - No need to search element by index.
  - Can represent structured data (like records) naturally.

Q12. Describe a scenario where using a tuple would be preferable over a list.
    When storing fixed data like coordinates (x, y) or employee code (sn001,st002) where modification is not needed.

Q13. How do sets handle duplicate values in Python?
   Sets automatically discard duplicates. For example, {1,2,2,3} becomes {1,2,3}.

Q14. How does the “in” keyword work differently for lists and dictionaries?
    In a list, in checks if the value exists and in a dictionary, in checks if the key exists (not the value).

Q15. Can you modify the elements of a tuple? Explain why or why not.
     No, tuples are immutable. Once created, their elements cannot be changed, added, or removed.

Q16. What is a nested dictionary, and give an example of its use case.
    A nested dictionary is a dictionary containing other dictionaries as values.
Example:
student = {“name” : “pooja” , “credentials” : { “class“ : 10 , “roll no” : 21}

Q17. Describe the time complexity of accessing elements in a dictionary.
   The average time complexity is O(1) (constant time) because of hash tables. In rare cases (hash collisions), it may degrade to O(n).

Q18. In what situations are lists preferred over dictionaries?
 When order matters.

   For sequential data processing (like queues, stacks).

Q19. Why are dictionaries considered unordered, and how does that affect data retrieval?
   Before Python 3.7, dictionaries were unordered. From 3.7 onwards, they preserve insertion order, but retrieval is still based on key lookups, not index positions.

Q20. Explain the difference between a list and a dictionary in terms of data retrieval.
   List: Retrieve by index (e.g., mylist[0]).
   Dictionary: Retrieve by key (e.g., mydict['name']).
